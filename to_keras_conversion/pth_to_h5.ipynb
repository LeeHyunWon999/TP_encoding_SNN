```python
import torch
import torch.nn as nn
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, BatchNormalization, ReLU, Dropout, InputLayer
from tensorflow.keras.layers import ZeroPadding1D, Conv1D, BatchNormalization, ReLU, AveragePooling1D, Flatten, Dense


pt_weights = {k: v.cpu().numpy() for k, v in weight.items()}


# Keras 모델 정의
def create_keras_model_MLP(num_class=2):
    model = Sequential([
        Input(shape=(187,), name='input'),
        Dense(128, input_shape=(187,), use_bias=False, name='fc'),
        ReLU(name='relu'),
        Dropout(0.5, name='dropout'),
        Dense(num_class, use_bias=False, name='out')
    ])
    return model

def create_keras_model_MLP_F(num_class=2):

    inputs = Input(shape=(187,), name='input')
    x = Dense(128, input_shape=(187,), use_bias=False, name='fc')(inputs)
    x = ReLU(name='relu')(x)
    x = Dropout(0.5, name='dropout')(x)
    prediction = Dense(num_class, use_bias=False, name='out')(x)

    return Model(inputs = inputs, outputs=prediction)

def create_keras_model_CNN(input_shape=(187, 1), num_class=2):
    model = Sequential([
        Input(shape=(187,1), name='input'),
        
        Conv1D(16, kernel_size=3, strides=1, padding='valid', activation='relu', input_shape=input_shape, name='conv1_conv', use_bias = False),
        ZeroPadding1D(padding=1, name='conv1_padding'),
        AveragePooling1D(pool_size=3, strides=2, padding='valid', name='conv1_pool'),

        Conv1D(64, kernel_size=3, strides=1, padding='valid', activation='relu', name='conv2_conv',use_bias = False),
        ZeroPadding1D(padding=1, name='conv2_padding'),
        AveragePooling1D(pool_size=2, strides=2, padding='valid', name='conv2_pool'),

        Conv1D(128, kernel_size=3, strides=1, padding='valid', activation='relu', name='conv3_conv', use_bias = False),
        ZeroPadding1D(padding=1, name='conv3_padding'),
        AveragePooling1D(pool_size=2, strides=2, padding='valid', name='conv3_pool'),

        Flatten(name='flatten'),
        Dense(500, activation='relu', name='fc1', use_bias = False),  
        Dense(num_class, name='fc2', use_bias = False)
    ])
    return model

def create_keras_model_CNN_F(input_shape=(187, 1), num_class=2):

    inputs = Input(shape=(187,1), name='input')
    
    x = Conv1D(16, kernel_size=3, strides=1, padding='valid', activation='relu', input_shape=input_shape, name='conv1_conv', use_bias = False)(inputs)
    x = ZeroPadding1D(padding=1, name='conv1_padding')(x)
    x = AveragePooling1D(pool_size=3, strides=2, padding='valid', name='conv1_pool')(x)

    x = Conv1D(64, kernel_size=3, strides=1, padding='valid', activation='relu', name='conv2_conv',use_bias = False)(x)
    x = ZeroPadding1D(padding=1, name='conv2_padding')(x)
    x = AveragePooling1D(pool_size=2, strides=2, padding='valid', name='conv2_pool')(x)

    x = Conv1D(128, kernel_size=3, strides=1, padding='valid', activation='relu', name='conv3_conv', use_bias = False)(x)
    x = ZeroPadding1D(padding=1, name='conv3_padding')(x)
    x = AveragePooling1D(pool_size=2, strides=2, padding='valid', name='conv3_pool')(x)

    x = Flatten(name='flatten')(x)
    x = Dense(500, activation='relu', name='fc1', use_bias = False)(x)
    prediction = Dense(num_class, name='fc2', use_bias = False)(x)

    return Model(inputs = inputs, outputs=prediction)



# MLP
# Keras 가중치 설정

# keras_model = create_keras_model_MLP_F( num_class=2)
# keras_model.get_layer('fc').set_weights([pt_weights['fc.weight'].T])
# keras_model.get_layer('out').set_weights([pt_weights['out.weight'].T])

# keras_model.save('/data/hongwonseok/ECG-SNN/KIST/MLP/MLP_binary_NO_BatchNorm_Input_F.h5')

# CNN
# #Keras 가중치 설정

keras_model = create_keras_model_CNN_F(num_class=2)

keras_model.get_layer(name='conv1_conv').set_weights([pt_weights['conv1.0.weight'].transpose(2, 1, 0)])


keras_model.get_layer(name='conv2_conv').set_weights([pt_weights['conv2.0.weight'].transpose(2, 1, 0)])

keras_model.get_layer(name='conv3_conv').set_weights([pt_weights['conv3.0.weight'].transpose(2, 1, 0)])


keras_model.get_layer(name='fc1').set_weights([pt_weights['fc.1.weight'].T])
keras_model.get_layer(name='fc2').set_weights([pt_weights['fc.3.weight'].T])




keras_model.save('/data/hongwonseok/ECG-SNN/KIST/CNN/CNN_binary_NO_BatchNorm_Input_F.h5')
```
