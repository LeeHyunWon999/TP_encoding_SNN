통계 파일로부터 저장된 csv 파일 불러다가 평균내는 역할.

```python
# 사전 변수
csv_dir = '/home/hschoi/leehyunwon/ECG-SNN/new_server/ver6/statistics/tensorboard_to_pandas/stride_2.csv'
savefile_name = 'stride_avg.csv'
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
# CSV 파일 불러오기
data = pd.read_csv(csv_dir)

# 폴더명에서 시간 기록(yyyy-mm-dd-hh-mm-ss)을 추출하여 그룹화 키 추가
def extract_time_key(source_dir):
    return "-".join(source_dir.split("_")[:-2])

data['time_key'] = data['source_dir'].apply(extract_time_key)

# metric에 해당하는 열만 추출 (acc, auroc, f1_score 같은 열 : stride 작업에선 source_dir만 빼면 될 듯?)
metric_data = data.drop(columns=['source_dir'])

print(metric_data)

# 시간 키 단위로 그룹화하여 평균과 표준편차 계산
grouped_metrics = metric_data.groupby('time_key').agg(['mean', 'std'])

print(grouped_metrics)
```

```python
# 그룹화 결과에 그룹화 키를 열로 복구
grouped_metrics_reset = grouped_metrics.reset_index()

# CSV로 저장
grouped_metrics_reset.to_csv(savefile_name, index=False)
```
