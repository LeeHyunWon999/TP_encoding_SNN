```python
import numpy as np

# 예시 1차원 배열 (적절한 숫자들이 이어지다가 200, 201, 202가 10번 반복되는 구조)
# data = np.array([1, 2, 3, 4, 200, 201, 202, 5, 6, 7, 8, 200, 201, 202] * 10)
# print(data)

data = log_spike

# 202 값의 인덱스를 기준으로 분할
indices = np.where(data == 202)[0] + 1  # 202 뒤에서 자르기 위해 +1
print(indices)

# 배열을 인덱스에 따라 나누기
split_arrays = np.split(data, indices)
print(split_arrays)

# 마지막 빈 배열 제거
split_arrays = split_arrays[:-1]

# 리스트로 변환하여 출력
split_arrays = [arr.tolist() for arr in split_arrays]

# 결과 출력
for i, arr in enumerate(split_arrays):
    print(f"리스트 {i + 1}({len(arr)}개): {arr}")
```

```python
# 200, 201, 202 제거
def remove_values(arr, values_to_remove):
    return [x for x in arr if x not in values_to_remove]

# 제거할 값들
values_to_remove = [200, 201, 202]

# 각 리스트에서 200, 201, 202 제거
cleaned_split_arrays = [remove_values(arr, values_to_remove) for arr in split_arrays]

# 결과 출력
for i, arr in enumerate(cleaned_split_arrays):
    print(f"리스트 {i + 1}({len(arr)}개): {arr}")
```

```python
# 입력값도 필요없으니 대충 900 이하의 값들을 제거
def remove_below_threshold(arr, threshold):
    return [x for x in arr if x > threshold]

# 900 이하의 값 제거 (threshold를 900으로 설정)
threshold = 900
filtered_split_arrays = [remove_below_threshold(arr, threshold) for arr in cleaned_split_arrays]

# 결과 출력
for i, arr in enumerate(filtered_split_arrays):
    print(f"리스트 {i + 1}({len(arr)}개): {arr}")
```

```python
# 2048, 2049도 제거한다. 이건 이제 2000 이상의 값을 제거하는 식으로 동작하도록.
def remove_above_threshold(arr, threshold):
    return [x for x in arr if x < threshold]

# 2000 이상의 값 제거
threshold = 2000
filtered_split_arrays_2 = [remove_above_threshold(arr, threshold) for arr in filtered_split_arrays]

# 결과 출력
for i, arr in enumerate(filtered_split_arrays_2):
    print(f"리스트 {i + 1}({len(arr)}개): {arr}")
```

```python
# 이제 각 리스트마다 어느 값이 몇번 찍혔는지를 보고 싶다.
from collections import Counter

# float 값을 int로 변환한 후, 각 리스트에서 값의 빈도수를 계산
counted_arrays = [Counter([int(x) for x in arr]) for arr in filtered_split_arrays_2]

# 결과 출력
for i, counted in enumerate(counted_arrays):
    print(f"리스트 {i + 1}: {counted}")
```

```python
# 이번엔 해당 값들을 128로 나눈다.

# 각 카운트 값을 128로 나누는 함수
def divide_counts_by_128(count_dict):
    return {key: count / 128 for key, count in count_dict.items()}

# 각 리스트에서 128로 나눈 값 계산
divided_counts_arrays = [divide_counts_by_128(counted) for counted in counted_arrays]

# 결과 출력
for i, divided in enumerate(divided_counts_arrays):
    print(f"리스트 {i + 1} (128로 나눈 값): {divided}")
```
